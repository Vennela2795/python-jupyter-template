### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# data
x = list(range(11))          # 0 → 10
y = [n**2 for n in x]        # square of each x

# line plot
plt.plot(x, y, marker='o')   # dots at each point

# customization
plt.title("y = x²")          # title
plt.xlabel("x")              # x-axis label
plt.ylabel("y")              # y-axis label
plt.grid(True)               # add grid

plt.show()
```
